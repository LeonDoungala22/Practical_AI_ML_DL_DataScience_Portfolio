# Decision Trees

[Doc sklearn](https://scikit-learn.org/stable/modules/tree.html)

**Decision Trees (DTs)** are a **non-parametric** supervised learning method used for **classification** and **regression**. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features. A tree can be seen as a piecewise constant approximation.

### - Advantages vs Disadvantages

| **Advantages of Decision Trees**                                       | **Disadvantages of Decision Trees**                                                                  |
|-----------------------------------------------------------------------|------------------------------------------------------------------------------------------------------|
| 1. Simple to understand and interpret. Trees can be visualized.       | 1. Decision-tree learners can create over-complex trees that do not generalize the data well (overfitting).|
| 2. Requires little data preparation; minimal need for data normalization. | 2. Decision trees can be unstable, generating different trees with small variations in the data.    |
| 3. Logarithmic cost for predicting data based on the number of data points.| 3. Predictions are piecewise constant and not smooth or continuous, making them less suitable for extrapolation.|
| 4. Handles both numerical and categorical data.                       | 4. Learning an optimal decision tree is NP-complete, leading to heuristic algorithms and potential bias.|
| 5. Can handle multi-output problems.                                  | 5. Hard to express certain concepts (e.g., XOR, parity) with decision trees.                           |
| 6. Uses a white box model, providing easily interpretable explanations.| 6. Bias in tree creation if some classes dominate; dataset balancing is recommended.                    |
| 7. Possible to validate a model using statistical tests.              |                                                                                                      |
| 8. Performs well even with some violations of assumptions.            |                                                                                                      |

### -  Classification

**DecisionTreeClassifier** is a class capable of performing multi-class classification on a dataset.

```python
from sklearn import tree

X = [[0, 0], [1, 1]]
Y = [0, 1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
clf.predict([[2., 2.]])
```

**NB:** In case that there are multiple classes with the same and **highest probability**, the classifier will predict the class with the lowest index amongst those classes.

```python
clf.predict_proba([[2., 2.]])
```

- **Example** 

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

- **Plot Tree** 

```python
tree.plot_tree(clf)
```

### Regression

Decision trees can also be applied to regression problems, using the **DecisionTreeRegressor** class.

```python
from sklearn import tree
X = [[0, 0], [2, 2]]
y = [0.5, 2.5]
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X, y)
clf.predict([[1, 1]])
```
