## 2. Data Cleaning
        
        -Check for missing values and outliers.
        -Ensure data consistency and correctness.

```python
#Missing values check 

df.isnull().sum()
```

```python
#Visualize missing value 

# missinggno is used to visualized missing values

import missingno as msno 

msno.matrix(df)
```

```python
df_CS_mean = df["Credit Score"].mean()
df_AI_Annual_Income_mean = df['Annual Income'].mean()
df_MSLD_mean = df["Months since last delinquent"].mean()
df_Y_Mode = df['Years in current job'].mode()[0]
df_B_mode = df['Bankruptcies'].mode()[0]

df["Credit Score"].fillna(df_CS_mean , inplace=True)
df["Annual Income"].fillna(df_AI_Annual_Income_mean , inplace=True)
df["Months since last delinquent"].fillna(df_MSLD_mean , inplace=True)
df["Years in current job"].fillna(df_Y_Mode , inplace=True)
df["Bankruptcies"].fillna(df_B_mode , inplace=True)
```

```python
#Remove id's
df.drop(columns=['Loan ID', 'Customer ID'], inplace=True)
```
