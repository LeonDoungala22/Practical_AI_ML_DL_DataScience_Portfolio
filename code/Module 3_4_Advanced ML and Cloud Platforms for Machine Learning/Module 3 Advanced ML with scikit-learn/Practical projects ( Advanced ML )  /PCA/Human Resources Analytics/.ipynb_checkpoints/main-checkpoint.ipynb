```python
#Features normalisation : Numerical features 

scaler = StandardScaler()
scaler_scaled = scaler.fit_transform(numerical_df)
scaler_scaled = pd.DataFrame(scaler_scaled , columns=numerical_df.columns)
scaler_scaled.head()
```

```python
# Instantiate the OneHotEncoder
encoder = OneHotEncoder()

# Fit and transform the categorical data
encoded_array = encoder.fit_transform(categorical_df).toarray()

# Create a DataFrame with column names
column_names = encoder.get_feature_names_out(categorical_df.columns)
encode_df = pd.DataFrame(encoded_array, columns=column_names)
encode_df
```

```python
# Merge scaler_scaled and encode_df
preprocessed_df = pd.merge(scaler_scaled, encode_df, how='inner', left_index=True, right_index=True)
preprocessed_df
```

```python
# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap
plt.figure(figsize=(18, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap of Features')
plt.show()
```

### 5. Visualization
- Visualize the dataset before and after PCA.
- Use scatter plots or other suitable visualizations.
___

- Visualize the dataset before and after PCA.

```python
X = preprocessed_df

# Assuming X has three features after preprocessing (change indices accordingly)
fig = plt.figure(figsize=(20, 8))

# Original Dataset subplot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
scatter1 = ax1.scatter(X.iloc[:, 0], X.iloc[:, 1], X.iloc[:, 2], c=X.iloc[:, 2], cmap='viridis', edgecolors='k', marker='o')
ax1.set_title('Original Dataset (3D)')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
ax1.set_zlabel('Feature 3')
fig.colorbar(scatter1, ax=ax1, label='Feature 3')

# PCA subplot
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

ax2 = fig.add_subplot(1, 2, 2)
scatter2 = ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=X.iloc[:, 2], cmap='viridis', edgecolors='k', marker='o')
ax2.set_title('Dataset After PCA')
ax2.set_xlabel('Principal Component 1')
ax2.set_ylabel('Principal Component 2')
fig.colorbar(scatter2, ax=ax2, label='Feature 3')

plt.tight_layout()
plt.show()
```

- Utilize the k-means algorithm to partition data into distinct clusters.

```python

# KMeans clustering with a range of k values
inertia = []
k_values = range(1, 11)  

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)

# Plotting the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

### 7. Cluster Visualization
- Implement cluster visualization to identify patterns and groupings within the dataset.
- Discuss the insights gained from cluster visualization.
____

## Number of cluseter = 4 

```python
# Corrected labels with more description
X = preprocessed_df

# KMeans clustering
kmeans = KMeans(n_clusters=4)  
labels = kmeans.fit_predict(X)

# After PCA with clustering colors
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', edgecolors='k', marker='o')
plt.title('Dataset After PCA with Clustering Colors')
plt.xlabel('Principal Component 1: Year or Duration Distribution')
plt.ylabel('Principal Component 2: Performance Indicator')

# Add legend
legend = plt.legend(*scatter.legend_elements(), title="Clusters")
plt.tight_layout()
plt.show()
```

### 6. Correlation Circle
- Add a Correlation Circle to visually represent the relationships between variables.
- Interpret the Correlation Circle to understand the relationships and contributions of variables in the reduced-dimensional space.
____

- Add a Correlation Circle to visually represent the relationships between variables. (Identify important features: top 5)

```python
# Identify important features based on loadings
feature_importance = np.abs(pca.components_)

# Choose a threshold for importance
threshold = 0.2

# Get the indices of the top 5 important features
top_5_indices = np.argsort(np.max(feature_importance, axis=0))[-5:]

# Extract loadings and variable names for the top 5 important features
top_5_loadings = pca.components_[:, top_5_indices].T * np.sqrt(pca.explained_variance_)
top_5_feature_names = preprocessed_df.columns[top_5_indices]

# Set up the figure
plt.figure(figsize=(10, 10))

# Plotting the loadings as vectors for the top 5 important features
for i, (x, y) in enumerate(zip(top_5_loadings[:, 0], top_5_loadings[:, 1])):
    plt.arrow(0, 0, x, y, color='r', alpha=0.8, width=0.01)
    plt.text(x, y, top_5_feature_names[i], color='g', ha='center', va='center')

# Plotting a unit circle
circle = plt.Circle((0, 0), 1, color='b', fill=False)  # Unit circle with radius 1
plt.gca().add_patch(circle)

# Set plot labels and title
plt.title('PCA Correlation Circle for Top 5 Important Features')
plt.xlabel('Principal Component 1 ( Year or Duration Distribution )')
plt.ylabel('Principal Component 2 ( Performance Indicator )')

# Set axis equal and grid
plt.axis('equal')
plt.grid(True)

# Show the plot
plt.show()
```

####  Discuss the insights gained from cluster visualization.

- **Cluster 0:** High Performance Rating and a substantial percentage of salary, but the distribution of years in the company is not significantly skewed towards the median compared to the overall employee year distribution.

- **Cluster 1:** High Performance Rating and a high percentage of salary with a linear distribution of years in the company.

- **Cluster 2:** Low Performance Rating and a low percentage of salary, yet the distribution of years in the company is notable.

- **Cluster 3:** Median Performance Rating and a median percentage of salary, with a distribution of years in the company that aligns with the overall employee year distribution.
