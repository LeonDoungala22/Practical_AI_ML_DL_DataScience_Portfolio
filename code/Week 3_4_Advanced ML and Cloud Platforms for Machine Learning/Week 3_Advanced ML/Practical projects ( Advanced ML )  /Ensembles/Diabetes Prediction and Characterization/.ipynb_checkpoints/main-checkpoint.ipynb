# Diabetes Types Overview

In the realm of diabetes, various types exist, each with distinct characteristics and risk factors.

## - Types of Diabetes

| Type | Description                                     | Characteristics                                     | Risk Factors                                       |
|:----:|:-----------------------------------------------:|:---------------------------------------------------:|:---------------------------------------------------:|
| **Type 1**        | Autoimmune disorder; immune system attacks insulin-producing cells. | - Diagnosed in childhood. <br> - Requires insulin injections. | - Genetic predisposition. <br> - Viral infections. |
| **Type 2**        | Insulin resistance, impaired production.          | - Usually in adulthood. <br> - Linked to poor lifestyle. | - Obesity. <br> - Family history. <br> - Age. |
| **Gestational**   | Develops during pregnancy; resolves post-childbirth. | - Elevated blood sugar during pregnancy. | - Overweight/obesity in pregnancy. <br> - Family history. |
| **LADA**          | Slow-onset autoimmune diabetes in adults.        | - Resembles Type 1 in adults. <br> - Progresses slowly. | - Autoimmune factors. <br> - Family history. |

***Note:** This table provides a concise summary of key information for each diabetes type.*

## 2. Feature Engineering and Data Preprossessing:
- Data Preprossessing

- Data Preprossessing

```python
#Separe Features and Target 

X = df.drop('Diabetes',axis=1)
y = df['Diabetes']
```

```python
# Initialize a StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled , columns=X.columns)
```

**NB:** 
- **StandardScaler** is less sensitive to outliers because it uses the mean and standard deviation for scaling.
- Outliers influence the mean more but have a smaller impact on the standard deviation, making StandardScaler more robust compared to methods relying on data range.

## 3. Ensemble Model Implementation:
- Implement ensemble models: Gradient Boosting, Random Forests, Bagging, Voting, and Stacking.
- Split the dataset into training and testing sets.
- Train and tune each ensemble model using appropriate hyperparameter tuning.
______

- Implement ensemble models: Gradient Boosting, Random Forests, Bagging, Voting, and Stacking.

```python
# Split Dataset into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Initialize individual models with some regularization
random_forest_model = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42)
gradient_boosting_model = GradientBoostingClassifier(n_estimators=100, max_depth=3, random_state=42)
xgboost_model = XGBClassifier(n_estimators=100, max_depth=3, learning_rate=0.1, reg_alpha=1e-5, reg_lambda=1e-5, random_state=42)
adaboost_model = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=100, learning_rate=1.0, random_state=42)
```

```python

# Train individual models with cross-validation
def train_and_evaluate(model, X_train, y_train):
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f"Cross-Validation Accuracy: {scores.mean():.3f} (+/- {scores.std() * 2:.3f})")

train_and_evaluate(random_forest_model, X_train, y_train)
train_and_evaluate(gradient_boosting_model, X_train, y_train)
train_and_evaluate(xgboost_model, X_train, y_train)
train_and_evaluate(adaboost_model, X_train, y_train)
```

```python
# Make predictions on the test set
random_forest_model.fit(X_train, y_train)
gradient_boosting_model.fit(X_train, y_train)
xgboost_model.fit(X_train, y_train)
adaboost_model.fit(X_train, y_train)
```

## 4. Model Evaluation and Comparison:
- Evaluate ensemble models using metrics such as accuracy, precision, recall, and F1-score.
- Visualize and compare feature importance across ensemble models.
- Analyze and compare the strengths and weaknesses of each ensemble technique.
_____

- Evaluate ensemble models using metrics such as accuracy, precision, recall, and F1-score.

```python
#Predict 

rf_predictions = random_forest_model.predict(X_test)
gb_predictions = gradient_boosting_model.predict(X_test)
xgboost_predictions = xgboost_model.predict(X_test)
adaboost_predictions = adaboost_model.predict(X_test)
```

```python
# Evaluate individual models
print("Random Forest Accuracy:", accuracy_score(y_test, rf_predictions))
print("Gradient Boosting Accuracy:", accuracy_score(y_test, gb_predictions))
print("XGBoost Accuracy:", accuracy_score(y_test, xgboost_predictions))
print("AdaBoost Accuracy:", accuracy_score(y_test, adaboost_predictions))

# Ensemble using VotingClassifier
voting_classifier = VotingClassifier(estimators=[
    ('random_forest', random_forest_model),
    ('gradient_boosting', gradient_boosting_model),
    ('xgboost', xgboost_model),
    ('adaboost', adaboost_model)
], voting='hard')

# Train the ensemble model with cross-validation
train_and_evaluate(voting_classifier, X_train, y_train)

# Make predictions on the test set
voting_classifier.fit(X_train, y_train)
voting_predictions = voting_classifier.predict(X_test)

# Evaluate VotingClassifier
print("Voting Classifier Accuracy:", accuracy_score(y_test, voting_predictions))
```

```python

DataViz.plot_learning_curve_with_overfitting_indicator(random_forest_model, X_train, y_train, 'Random Forest Learning Curve')
DataViz.plot_learning_curve_with_overfitting_indicator(gradient_boosting_model, X_train, y_train, 'Gradient Boosting Learning Curve')
DataViz.plot_learning_curve_with_overfitting_indicator(xgboost_model, X_train, y_train, 'XGBoost Learning Curve')
DataViz.plot_learning_curve_with_overfitting_indicator(adaboost_model, X_train, y_train, 'AdaBoost Learning Curve')
```

 - Hyperparameter tuning : XGBoost as the best model

```python


# XGBoost hyperparameters
xgboost_params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_child_weight': [1, 3, 5]
}

# Create GridSearchCV instance for XGBoost
xgboost_grid_search = GridSearchCV(XGBClassifier(), xgboost_params, cv=5, scoring='accuracy')

# Fit the XGBoost model with the best hyperparameters
xgboost_grid_search.fit(X_train, y_train)

# Print the best hyperparameters and accuracy score for XGBoost
print("XGBoost Best Parameters:", xgboost_grid_search.best_params_)
print("XGBoost Best Accuracy:", xgboost_grid_search.best_score_)

# Use the best XGBoost model from the grid search
best_xgboost_model = xgboost_grid_search.best_estimator_
```

**NB:** We observe that the current tuned model demonstrates performance comparable to that of the original model.

- Visualize and compare feature importance across ensemble models.

```python
# Function to plot feature importance for XGBoost model
def plot_feature_importance(model, feature_names):
    importances = model.feature_importances_
    indices = np.argsort(importances)[::-1][:10]   #top 10

    plt.figure(figsize=(12, 8))  
    plt.bar(range(10), importances[indices], align="center")
    plt.xticks(range(10), feature_names[indices], rotation=45, ha="right")
    plt.title('XGBoost - Feature Importance')
    plt.tight_layout() 

# Call the function to plot feature importance for XGBoost
plot_feature_importance(xgboost_model, X_train.columns)
```
