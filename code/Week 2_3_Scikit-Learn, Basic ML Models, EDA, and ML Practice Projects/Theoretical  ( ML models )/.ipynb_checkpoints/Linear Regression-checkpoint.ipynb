## Linear Models

- **Definition:** A linear model is a mathematical representation assuming a linear relationship between variables.

- **Equation:** 
  $$[ Y = \beta_0 + \beta_1X + \epsilon ]$$
  - \( Y \): Dependent variable
  - \( X \): Independent variable
  - \( \beta_0 \): Intercept
  - \( \beta_1 \): Slope
  - \( \epsilon \): Error term

- **Purpose:** The model aims to describe and predict relationships between variables based on a linear assumption.

- **Extensions:** Multiple linear regression is an extension that accommodates multiple independent variables.

- **Applications:** Commonly used in statistics, economics, machine learning, and various other fields for modeling and prediction.


## 1- LinearRegression 

- **LinearRegression** will take in its fit method arrays **X**, **y** and will store the coefficients 
 of the linear model in its **coef_** member. 
 
- fits a linear model with coefficients to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation. Mathematically it solves a problem of the form:
 
     $$min_w || X_w - y ||^2_2$$ 

## $$ R^2 $$

R-Squared (R or the coefficient of determination) is a statistical measure in a regression model that determines the proportion of variance in the dependent variable that can be explained by the independent variable. In other words, r-squared shows how well the data fit the regression model (the goodness of fit).

```python
from sklearn import linear_model
```

```python
linearR = linear_model.LinearRegression()
linearR.fit([[0, 0], [1, 1], [2, 2]], [0, 1, 2])
```
