```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Pyplot

#Draw a line in a diagram from position (0,0) to position (6,250):

x = np.array([0,6]) 
y = np.array([0,250])

plt.plot(x,y)
plt.show()
```

```python
#Plotting Without Line

#plot a diagram from position : (2,5) to position ( 8,20) , wihtout linge

x = np.array([2,8])
y = np.array([3,10])

plt.plot(x,y , 'o')
plt.show()
```

```python
#Multiple Points

#Plot the ligne diagram from position (3,10 ) to position ( 4,6 ) then from position (6,2) to position (7,10)

x = np.array([3,4,6,7])
y = np.array([10,6,2,10])

plt.plot(x,y)
plt.show()
```

```python
# Default X-Points

y = np.array([2,4,10,2,5,9,3,4,8,11])

plt.plot(y) # default x-point use 1,2,3,4,5,6 ... x - axis points by default
plt.show()
```

```python
#Matplotlib Markers

plt.plot(y,marker='o')
plt.show()
```

```python
# Format Strings fmt       # Usage : marker|line|color

y = np.array([3,4,5,2,7,9,12,6,16,0,-2])

plt.plot(y, 'o:r')
plt.show()
```

```python
# Format Strings fmt   for 2 graphs

y = np.array([3,4,5,2,7,9,12,6,16,0,-2])
y2 = np.array([3,5,5,6,7,7,12,4,17,0,-2])

plt.plot(y, 'o:r')
plt.plot(y2, 'o:b')
plt.show()
```

```python
# Marker Size

y = np.array([1,5,2,8,2,9,3,4,7])
plt.plot(y, marker='o',ms=10)
plt.show()
```

```python
#Marker Color

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 8, mec = 'r')
plt.show()
```

```python
#Matplotlib Line

plt.plot(y,linestyle='dotted')
plt.show()
```

```python
# Matplotlib Labels and Title

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.xlabel("Average Pulse")
plt.ylabel("Colorie Burnage")
plt.title("Sports whatch Data")
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
#Add Grid Lines to a Plot

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#Matplotlib Scatter

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#Compare Plots

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
#How to Use the ColorMap

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
#Matplotlib Bars

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```
