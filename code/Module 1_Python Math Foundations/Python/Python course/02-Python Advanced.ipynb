### - Python Lambda

A lambda function is a concise and unnamed function.
It has the flexibility to accept multiple arguments, yet it must consist of a single expression.

**Syntax :**


lambda arguments : expression

```python
#Example :

x = lambda a : a +1 

print(x(1))
```

```python
# Example 2 : 

y = lambda a,b : a * b 

print(y(3,2))
```

 #### - Why Use Lambda Functions?


The power of lambda is better shown when you use them as an anonymous function inside another function.

Say you have a function definition that takes one argument, and that argument will be multiplied with an unknown number:

```python
def myfunc(n):
    return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```

#### Example:

```python
# Define a class named "Person"
class Person:
    # The constructor method with four parameters: name, age, profession, and email
    def __init__(self, name, age, profession, email):
        # Initialize the class attributes
        self.name = name          # Attribute to store the person's name
        self.age = age            # Attribute to store the person's age
        self.profession = profession  # Attribute to store the person's profession
        self.email = email        # Attribute to store the person's email

    # Method to greet and return a message
    def greet(self):
        return f"Hello, my name is {self.name}, I am {self.age} years old, and I work as a {self.profession}."

    # Method to send an email
    def send_email(self, recipient, message):
        return f"Sent an email to {recipient} at {self.email}: {message}"
```

```python

# Create an object (instance) of the "Person" class
person1 = Person("Alice", 30, "Software Engineer", "alice@email.com")

# Access and print the "name" attribute of the object
print("Name:", person1.name)  # Output: "Name: Alice"

# Access and print the "age" attribute of the object
print("Age:", person1.age)    # Output: "Age: 30"

# Access and print the "profession" attribute of the object
print("Profession:", person1.profession)  # Output: "Profession: Software Engineer"

# Access and print the "email" attribute of the object
print("Email:", person1.email)  # Output: "Email: alice@email.com"

# Call the "greet" method of the object to get a greeting message
greeting = person1.greet()
print(greeting)  # Output: "Hello, my name is Alice, I am 30 years old, and I work as a Software Engineer."

# Use the "send_email" method to send an email
email_message = person1.send_email("Bob", "Hello, Bob! How are you?")
print(email_message)  # Output: "Sent an email to Bob at alice@email.com: Hello, Bob! How are you?"
```
